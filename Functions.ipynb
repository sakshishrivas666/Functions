1.  What is the difference between a function and a method in Python ?
    - A function is a block of code that performs a specific task and can be called independently.

    example:-     
              def greet(name):   # Function definition
                    print(f"Hello, {name}!")

    example:-

             def greet(self):  # Method definition
                  print("Hello!")

        p = Person()
        p.greet()            # Calling the method on the object

      - Key Difference: Methods are functions inside classes, called using objects. Functions can be called independently.

5.  What are iterators in Python and how do they differ from iterables?
    -  Iterable :- An iterable is any Python object that can return an iterator. It is an object that you can iterate (loop) over.

        It implements the __iter__() method.

       Examples: lists, tuples, strings, dictionaries, sets, etc.
      
       my_list = [1, 2, 3, 4]  #it is a iterable

    - Iterator :- An iterator is an object that represents a stream of data. It allows you to iterate through all the elements one by one.

     It implements both:

      __iter__() → returns the iterator object itself.

     __next__() → returns the next element or raises StopIteration when there are no more items.

     example:-

      
          # Get an iterator from the iterable
                 it = iter(my_list)

          # Iterate using next()
                 print(next(it))   # 10
    

.6 . Explain the concept of generators in Python and how they are defined
     
     
  # concept of generator

  n Python, a generator is a special type of function that generates a
  sequence of values using the yield keyword instead of return.
  Unlike regular functions that execute and return a single value,
  generators produce a series of values one at a time,
  pausing their execution after each yield and resuming when the
  next value is requested.
  This behavior makes them memory-efficient, especially when dealing
  with large datasets,
  as they avoid storing the entire sequence in memory at once.

   
   # defining Generators

  - Generators are typically defined like regular functions but
    
      with the yield keyword:

    example:-

             def my_generator(n):
                   for i in range(n):
                        yield i  # Pauses and yields the current value of i
                        
            # Create a generator object
             gen = my_generator(3)  

             # Iterate and print values
               print(next(gen))  # Output: 0
               print(next(gen))  # Output: 1
               print(next(gen))  # Output: 2

7.   What are the advantages of using generators over regular functions ?
    - # Advantages of Generators over Regular Functions

    - Memory Efficiency:-  Generators produce values one at a time instead of storing the entire sequence in memory. This is especially beneficial when dealing with large datasets, as it significantly reduces memory consumption.

    - Lazy Evaluation:-  Generators evaluate values only when requested. This "lazy" behavior allows them to handle infinite sequences or data streams that would be impossible to store in memory entirely.

    - mproved Performance:-  Generators can be faster in certain scenarios, particularly when dealing with large or complex data structures, as they avoid unnecessary computations until the values are needed.

    - Readability:-  Generators can make code more concise and readable, especially for tasks that involve iterating over data.

    example:-

   
         def even_numbers(n):
              for i in range(2, n + 1, 2):
                  yield i

          # Create a generator object
          gen = even_numbers(10)

          # Iterate and print values
            for num in gen:
                print(num)
              
              #output
              2
              4
              6
              8
              10

.9 . Explain the purpose and usage of the `map()` function in Python.


  # Purpose:-
The map() function applies a given function to each item of an iterable (like a list, tuple, etc.) and returns an iterator of the results.

 # Usage :-
 function:  The function to apply to each item.

 iterable:   The iterable containing the items to process.

  # syntax
  map(function, iterable)

#example

       numbers = [1, 2, 3, 4, 5]
       squared_numbers = list(map(lambda x: x**2, numbers))  
      print(squared_numbers)  # Output: [1, 4, 9, 16, 25]

  

10.  What is the difference between `map()`, `reduce()`, and `filter()` functions in Python ?

    - # map ():-
    Applies a function to each item in an iterable (like a list) and returns a new iterable with the results. Think of it as transforming each element individually.

    example:-

         numbers = [1, 2, 3, 4, 5]
         squared_numbers = list(map(lambda x: x**2, numbers))  
         print(squared_numbers)  # Output: [1, 4, 9, 16, 25]
# filter():-
  Filters elements from an iterable based on a condition specified by a function. It returns a new iterable containing only the elements that satisfy the condition. Think of it as picking out specific elements.

example:-

      numbers = [1, 2, 3, 4, 5]
      even_numbers = list(filter(lambda x: x % 2 == 0, numbers))  
      print(even_numbers)  # Output: [2, 4]

# reduce():-
  Applies a function cumulatively to the items of an iterable, reducing them to a single value. Think of it as combining elements step-by-step.

example:-

    from functools import reduce

    numbers = [1, 2, 3, 4, 5]
    product = reduce(lambda x, y: x * y, numbers)  
    print(product)  # Output: 120

```python
# Q.4 . Write a Python function that checks if a given number is prime or not from 1 to 200.
def is_prime(n):
  if n<=1:
    return False
  for i in range(2,n):
    if n%i == 0:
      return False
  return True

for num in range(1, 201):
  if is_prime(num):
    print(f"{num} is prime")
  else:
    print(f"{num} is not prime")
```

```python
# Q.10 . Create a Python program that uses `filter()` to remove all the vowels from a given string.
str_1 = "sakshi shrivas, i am from m.p, and this course (data analytics) is very interesting"

vowels = "aeiouAEIOU"
result = "".join(filter(lambda char: char not in vowels, str_1))
print(f"string without vowel: {result}")
```

```python
# # Q.11 ) Imagine an accounting routine used in a book shop. It works on a list with sublists, which look like this:
#          Write a Python program, which returns a list with 2-tuples. Each tuple consists of the order number and the
#         product of the price per item and the quantity. The product should be increased by 10,- € if the value of the
#         order is smaller than 100,00 €.
#         Write a Python program using lambda and map



# Ans:- Sample list with sublists: [Order Number, Price per item, Quantity]
orders = [
    [34587, 40.95, 4],   # 40.95 * 4 = 163.8 no increase
    [98762, 56.80, 5],   # 56.80 * 5 = 284 no increase
    [77226, 32.95, 3],   # 32.95 * 3 = 98.85 → +10 → 108.85
    [88112, 24.99, 3]    # 24.99 * 3 = 74.97 → +10 → 84.97
]

# Use lambda and map to create the result list
result = list(map(lambda order: (
    order[0],  # Order number
    (order[1] * order[2]) + (10 if order[1] * order[2] < 100 else 0)
       ) , orders))

# Print the result
print(result)
```
